```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
# Normalizing the image to [0,1] range

x_train = x_train/255
x_test = x_test/255
```

```python
x_train
```

```python
x_train[0]
```

```python
x_train.shape
```

```python
plt.matshow(x_train[0])
```

```python
for i in range (0,10):
    plt.imshow(x_train[i])
    plt.title(y_train[i])
    plt.axis('off')
    plt.show()
```

```python
x_train_flattened = x_train.reshape(len(x_train),28*28)
x_train_flattened
# x_train_flattened.shape
```

```python
x_train_flattened.shape
```

```python
x_test_flattened = x_test.reshape(len(x_test),28*28)
x_test_flattened.shape
```

```python
model = keras.Sequential([keras.layers.Dense(10,input_shape=(784,),activation = 'sigmoid')])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics =['accuracy'])
```

```python
model.fit(x_train_flattened, y_train,epochs=5)
```

```python
model.evaluate(x_test_flattened,y_test)
```

```python
y_pred = model.predict(x_test_flattened)
```

```python
plt.matshow(x_test[0])
```

```python
y_pred[0]
```

```python
np.argmax(y_pred[0])
```

```python
y_pred_lables = [np.argmax(i) for i in y_pred]
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions = y_pred_lables)
cm
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('truth')
```

```python
model2 = keras.Sequential([keras.layers.Dense(100,input_shape=(784,),activation = 'relu'),
                           keras.Sequential([keras.layers.Dense(10, activation = 'sigmoid')])])

model2.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics =['accuracy'])
model2.fit(x_train_flattened, y_train,epochs=5)
```

```python
model2.evaluate(x_test_flattened,y_test)
```

```python
y_pred2 = model2.predict(x_test_flattened)
```
